```
spark.read.format("csv")\
     .option("header", "true")\
     .option("inferSchema", "true")\
     .option("delimiter", ";")\
     .load("/FileStore/tables/arquivo_geral.csv")\
     .createOrReplaceTempView("tabela_covid")
```

```
%sql
DESCRIBE tabela_covid
```

Retornar a soma de casos novos por região, ordenando de acordo com a soma

```
%sql
SELECT regiao, SUM(casosNovos) AS SOMA_CASOS_NOVOS FROM tabela_covid GROUP BY regiao ORDER BY SOMA_CASOS_NOVOS
```

Retornar a soma de casos novos, o valor mínimo de casos novos, o valor máximo de casos novos, a média de casos novos ordenados por região

```
%sql
SELECT regiao, SUM(casosNovos) AS SOMA_CASOS, MIN(casosNovos) AS MIN_CASOS, MAX(casosNovos) AS MAX_CASOS, ROUND(MEAN(casosNovos), 2) AS MEDIA_CASOS
FROM tabela_covid 
GROUP BY regiao 
ORDER BY regiao
```
