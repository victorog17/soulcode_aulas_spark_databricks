```
# File location and type
#file_location = "/FileStore/tables/arquivo_geral.csv"
#file_type = "csv"

# CSV options
#infer_schema = "true"
#first_row_is_header = "true"
#delimiter = ";"

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format("csv") \
  .option("inferSchema", "true") \
  .option("header", "true") \
  .option("sep", ";") \
  .load("/FileStore/tables/arquivo_geral.csv") \
  .createOrReplaceTempView("tabela_covid")
```

```
%sql
SELECT * FROM tabela_covid;
```

```
%sql
SELECT regiao, estado, casosNovos FROM tabela_covid WHERE casosNovos > 10
```

Fazer a contagem de casos novos por região

```
%sql
SELECT regiao, COUNT(casosNovos) AS QTD_CASOS_NOVOS FROM tabela_covid GROUP BY regiao
```

Fazer a contagem de casos novos por região onde a quantidade for maior que 500

```
%sql
SELECT regiao, COUNT(casosNovos) AS QTD_CASOS_NOVOS 
FROM tabela_covid 
GROUP BY regiao 
HAVING QTD_CASOS_NOVOS > 500
```

Fazer a contagem de casos novos na região Nordeste

```
%sql
SELECT regiao, COUNT(casosNovos) AS QTD_CASOS_NOVOS FROM tabela_covid WHERE regiao == 'Nordeste' GROUP BY regiao
```
