```
df = spark.read.format("csv") \
  .option("inferSchema", "true") \
  .option("header", "true") \
  .option("sep", ";") \
  .load("/FileStore/tables/arquivo_geral.csv") \
  .createOrReplaceTempView("tabela_covid")
```

```
%sql
SELECT * FROM tabela_covid
```

```
%sql
SELECT * FROM tabela_covid WHERE estado LIKE 'R%'
```

Mostrar todos os dados da tabela covid da região norte, apenas dos estados AM e RO

```
%sql
SELECT * FROM tabela_covid WHERE regiao == 'Norte' AND (estado == 'AM' OR estado == 'RO')
```

```
spark.sql("SELECT estado FROM tabela_covid").display()
```

Selecionar todos os estados removendo a repetição

```
%sql
SELECT DISTINCT estado FROM tabela_covid
```

```
spark.sql("SELECT estado FROM tabela_covid").distinct().display()
```

Mostrar os dados referentes aos estados de RR, AC e RN

```
%sql
SELECT * FROM tabela_covid WHERE (estado = 'RR' OR estado = 'AC' OR estado = 'RN')
```

```
%sql
SELECT * FROM tabela_covid WHERE estado in ('RR', 'AC', 'RN')
```

Mostrar a quantidade de casos dos estados por região onde a quantidade for menor que 500 e ordenando por região de maneira decrescente

```
%sql
SELECT regiao, COUNT(casosNovos) AS QTD_CASOS FROM tabela_covid GROUP BY regiao HAVING QTD_CASOS < 500 ORDER BY regiao DESC LIMIT 2
```
