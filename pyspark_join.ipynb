O Pyspark Join é usado para combinar dois DataFrames e, encadeando-os, você pode juntar vários DataFrames;
Ele suporta todas as operações básicas de tipo de junção disponíveis no SQL tradicional, como INNER< LEFT OUTER, RIGHT OUTER, LEFT ANTI, LEFT SEMI, CROSS, SELF JOIN.

EXEMPLOS:


inner ==> INNER JOIN SQL

outer, full, fullouter, full_outer ==> FULL OUTER JOIN

left, leftouter, left_outer ==> LEFT JOIN

right, rightouter, right_outer ==> RIGHT JOIN

cross

anti, leftanti, left_anti

semi, leftsemi, left_semi

**BIBLIOTECAS**

```
import pyspark.sql.functions as F
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType
```

**CRIANDO UM DATAFRAME PYSPARK**

```
dados = [
         (1, "Anderson", 1000.00),
         (2, "Kennedy", 2000.00),
         (3, "Bruno", 2300.00),
         (4, "Maria", 2300.00),
         (5, "Eduardo", 2400.00),
         (6, "Mendes", 1900.00),
         (7, "Kethlyn", 1500.00),
         (8, "Thiago", 1800.00),
         (9, "Carla", 2100.00)
]

schema = ["id", "Nome", "Salario"]
df1 = spark.createDataFrame(data=dados, schema=schema)
df1.printSchema()
#df1.show()
display(df1)
```

```
dados2 = [
    (1, "Delhi", "India"),
    (2, "Tamil Nadu", "India"),
    (3, "London", "UK"),
    (4, "Sydney", "Australia"),
    (8, "New York", "USA"),
    (9, "California", "USA"),
    (10, "New Jersey", "USA"),
    (11, "Texas", "USA"),
    (12, "Chicago", "USA")
]
schema2 = ["id", "local", "pais"]
df2 = spark.createDataFrame(data=dados2, schema=schema2)
df2.printSchema()
display(df2)
```

**INNER JOIN**

Ele verifica algum campo em comum nos dois dataframes e retorna os resultados

```
df_inner = df1.join(df2, on=['id'], how='inner')
df_inner.display()
```

**OUTER JOIN**

Ele retorna todos os dados do dataframe da coluna da esquerda e direita

```
df_outer = df1.join(df2, on=['id'], how='outer')
df_outer.display()
```

**LEFT JOIN**

Ele retorna todos os dados da coluna da esquerda e os registros que combinam (fazem match)

```
df_left = df1.join(df2, on=['id'], how='left')
df_left.display()
```

**RIGHT JOIN**

Ele retorna todos os dados da coluna da direita e os registros que combinam (fazem match)

```
df_right = df1.join(df2, on=['id'], how='right')
df_right.display()
```

**ANTI LEFT JOIN**

Ele retorna todos os dados do dataframe da esquerda que não estão presentes no dataframe da direita

```
df_left_anti = df1.join(df2, on=['id'], how='left_anti')
df_left_anti.display()
```

**LEFT SEMI JOIN**

Retorna todos os elementos do dataframe da esquerda que possuam correspondencia com o dataframe da direita (bem parecido com inner join)

```
df_left_semi = df1.join(df2, on=['id'], how='left_semi')
df_left_semi.display()
```

**FULL JOIN**

Faz a mesma coisa do Outer Join

```
df_full = df1.join(df2, on=['id'], how='full')
df_full.display()
```

**ANTI JOIN**

Traz todos que não tem correspondencia

```
df_anti = df1.join(df2, on=['id'], how='anti')
df_anti.display()
```
