**BIBLIOTECAS**

```
import pyspark.sql.functions as F
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType
```

**CRIANDO UM DATAFRAME PYSPARK**

```
dados = [
         ("Anderson", "Vendas", "SP", 1500.00, 34, 1000.00),
         ("Kennedy", "Vendas", "CE", 1200.00, 56, 2000.00),
         ("Bruno", "Vendas", "SP", 1100.00, 30, 2300.00),
         ("Maria", "Financas", "CE", 3600.00, 24, 2300.00),
         ("Eduardo", "Financas", "CE", 4500.00, 40, 2400.00),
         ("Mendes", "Financas", "RS", 8000.00, 36, 1900.00),
         ("Kethlyn", "Financas", "RS", 1200.00, 53, 1500.00),
         ("Thiago", "Marketing", "GO", 1100.00, 25, 1800.00),
         ("Carla", "Marketing", "GO", 2600.00, 50, 2100.00)
]

schema = ["Nome", "Departamento", "Estado", "Salario", "Idade", "Bonus"]
df = spark.createDataFrame(data=dados, schema=schema)
df.printSchema()
#df.show()
display(df)
```

**approx_count_distinct Aggregate Function**

Exibe os valores distintos em um grupo dentro de um dataframe (nesse caso o grupo é o salário)

```
approx = df.select(F.approx_count_distinct("Salario")).collect()[0][0]
print(f"approx_count_distinct: {approx}")
```

**AVG Aggregate Function**

Retorna a média de valores através da coluna salário

```
avg = df.select(F.avg("Salario")).collect()[0][0]
print(f"avg: {avg}")
```

**collect_list Aggregate Function**

Pega todos os salários e coloca dentro de uma lista

```
df.select(F.collect_list("Salario")).show(truncate=False)
#df.select(F.collect_list("Salario")).display()
```

**collect_set Aggregate Function**

Retorna uma lista de valores distintos (salários)

```
df1 = df.select(F.collect_set("Salario"))
df1.show(truncate=False)
```

**CountDistinct Aggregate Function**

Retorna o número de elementos distintos

```
df2 = df.select(F.countDistinct(F.col("Departamento"), F.col("Salario")))
df2.show(truncate=False)

df.show()
```

**first function**

Retorna o primeiro valor

```
df.select(F.first(F.col("Salario"))).show(truncate=False)
```

**MAX**

Retorna o maior valor

```
df.select(F.max(F.col("Salario"))).show(truncate=False)
```

**MIN**

Retorna o menor valor

```
df.select(F.min(F.col("Salario"))).show(truncate=False)
```

**SUM**

Retorna a soma

```
df.select(F.sum(F.col("Salario"))).show(truncate=False)
```

**COUNT E COLLECT()[0][0]**

```
df.select(F.count(F.col("Salario"))).show(truncate=False)

valor = df.select(F.count("Salario")).collect()[0][0]
print(valor)
```
