```
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
```

```
df = spark.read.format("csv") \
  .option("inferSchema", "true") \
  .option("header", "true") \
  .option("sep", ",") \
  .load("/FileStore/tables/titanic_train.csv") \
  .createOrReplaceTempView("titanic")
```

```
%sql
SELECT * FROM titanic;
```

01.Verificando quantos homens com menos de 30 anos sobreviveram

```
%sql
SELECT PassengerId, Survived, Pclass, Name, Sex, Age 
FROM titanic 
WHERE Survived = 1 AND Age < 30 AND Sex = 'male';
```

02.Verificando número de sobreviventes por sexo

```
%sql
SELECT Sex, COUNT(Survived) AS Survived_by_sex
FROM titanic
WHERE Survived = 1
GROUP BY Sex
```

03.Verificando número de sobreviventes por classe

```
%sql
SELECT Pclass, COUNT(Survived) AS Survived_by_class
FROM titanic
WHERE Survived = 1
GROUP BY Pclass
```

04.Verificando quantos homens que estavam na terceira classe não sobreviveram

```
%sql
SELECT * FROM titanic WHERE (Survived = 0 AND Sex = 'male' AND Pclass = 3)
```

05.Consulta de quantos passageiros tinham mais de 50 anos e não sobreviveram em ordem decrescente

```
%sql
SELECT * FROM titanic  WHERE Age > 50 AND Survived = 0 ORDER BY Age DESC
```

06.Verificando Quantidade de sobreviventes e não sobreviventes de ambos os sexos

```
%sql
SELECT Survived, Sex, COUNT(Sex) AS N_Sex
FROM titanic
GROUP BY Survived, Sex
ORDER BY Sex, Survived
```

07.Consulta de quantidade de sobreviventes e não sobreviventes por classe e sexo

```
%sql
SELECT Survived, Sex, Pclass, COUNT(Sex) AS N_Sex
FROM titanic
GROUP BY Survived, Sex, Pclass
ORDER BY Survived DESC, Pclass, N_Sex DESC
```

08.Consulta de sobreviventes maior de 30 anos por sexo

```
%sql
SELECT Survived, Sex, COUNT(Age) AS Maior_30
FROM titanic
WHERE Age > 30
GROUP BY Survived, Sex
ORDER BY Survived DESC, Sex
```

09.Distribuição da idada dos passageiros, menor, maior e média por sexo

```
%sql
SELECT Sex, MIN(Age) AS menor_idade, MAX(Age) AS maior_idade, ROUND(MEAN(Age), 2) AS media_idade 
FROM titanic
GROUP BY Sex
```

10.Distribuição de Idades de acordo com sexo e classe

```
%sql
SELECT Sex, Pclass, MIN(Age) AS menor_idade, MAX(Age) AS maior_idade, ROUND(MEAN(Age), 2) AS media_idade 
FROM titanic
GROUP BY Pclass, Sex
ORDER BY Pclass, Sex
```
