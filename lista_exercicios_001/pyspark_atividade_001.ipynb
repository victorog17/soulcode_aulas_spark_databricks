**BIBLIOTECAS**

```
import pyspark.sql.functions as F
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType
from pyspark.sql.window import Window
```

```
dados = [
    (1, 'Bob', 45, 'M'),
    (2, 'Gloria', 43, 'F'),
    (4, 'Albert', 28, 'M'),
    (5, 'Laura', 33, 'F'),
    (8, 'Simone', 18, 'T'),
    (12, 'Marta', 45, 'F'),
    (45, 'Jairo', 82, 'M'),
    (13, 'Teste', 38, 'T')
]

schema = ['id', 'nome', 'idade', 'gen']
df = spark.createDataFrame(data=dados, schema=schema)
df.printSchema()
df.display()
```

1) Verifique as pessoas com idade superior a 30 anos

```
df.filter(F.col("idade") > 30).display()
```

2) Verificar apenas as pessoas do sexo F

```
df.filter(F.col("gen") == 'F').display()
```

3) Verifique a quantidade de pessoas por sexo

```
display(df.groupby(F.col("gen")).count())
```

4) Calcular a m√©dia de idade das pessoas

```
df.groupby().mean("idade").display()
```

5) Fazer uma sele√ß√£o apenas do nome e idade das pessoas ordenando por nome

```
df.select(F.col("nome"), F.col("idade")).orderBy("nome").display()
```

6) Mostrar um ranking de pessoas por idade (Do mais novo para o mais velho)

```
w0 = Window.orderBy("idade")
df2 = df.withColumn("rank", F.row_number().over(w0))
df2.show()
```
